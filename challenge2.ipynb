```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('company_sales_data.csv')
data
```

```python
data1=data.loc[data['facewash']<1500][['month','facecream','facewash']]
data1
```

```python
X_axis = int(data1['month'].count())
plt.figure(figsize=(11,8))
plt.rcParams.update({'font.size': 15})
plot=plt.subplot(1,1,1)
plot.bar(data1['month'], data1['facewash'],width=0.4,color='#E3CF57', label = 'Face wash sales',align='center')
plot.bar(data1['month'], data1['facecream'],width=0.2,color='#A52A2A', label = 'Face Cream sales',align='edge')


plt.xlabel("Months")
plt.ylabel("Sales units in number")
plt.suptitle('Facewash sales VS Facecream sales', fontsize=18, fontweight='bold',y=0.95)
plt.title("For the months with facewash sales less than 1500")
plt.grid(color = '#B7B6B5', linestyle = '--', linewidth = 1.5)
plt.legend()
plt.tight_layout()
plt.show()
plt.savefig('hi')
```
